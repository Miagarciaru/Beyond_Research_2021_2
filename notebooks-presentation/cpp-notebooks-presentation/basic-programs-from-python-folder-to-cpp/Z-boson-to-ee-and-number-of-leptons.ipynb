```c++
#include <TChain.h>
#include <vector>
#include <TFile.h>
#include <iostream>
#include <string>
#include <stdio.h>
```

```c++
TFile *file = TFile::Open("https://atlas-opendata.web.cern.ch/atlas-opendata/samples/2020/1largeRjet1lep/MC/mc_361106.Zee.1largeRjet1lep.root");
```

```c++
TCanvas *canvas = new TCanvas();
```

```c++
TTree *tree = (TTree*) file->Get("mini");
int nentries = tree->GetEntries();
cout << nentries << endl;
```

```c++
UInt_t lep_n = -1;
tree->SetBranchAddress("lep_n", &lep_n);
```

```c++
TH1F *hist = new TH1F("variable","Example plot: Number of leptons; Number of leptons ; Events ",5, 0, 5);
```

```c++
for (int ii=0; ii<nentries; ii++)
{
    tree->GetEntry(ii);
    hist->Fill(lep_n);
}
cout << "Done!" << endl;
```

```c++
hist->SetLineColor(4);
hist->SetFillStyle(3001);
hist->SetFillColor(kBlue);
hist->DrawNormalized("");
canvas->Draw();
```
